# Operator and Method Overloading

__Operator  Overloading__ berarti memberikan makna yang diperluas di luar makna operasional yang telah ditentukan sebelumnya. Misalnya operator digunakan untuk menambahkan dua bilangan bulat serta menggabungkan dua string dan menggabungkan dua daftar. Ini dapat dicapai karena operator + kelebihan beban oleh kelas int dan kelas str. Anda mungkin telah memperhatikan bahwa operator atau fungsi bawaan yang sama menunjukkan perilaku yang berbeda untuk kelas yang berbeda, ini disebut Operator Overloading.

Bagaimana cara membebani operator dengan Python?
Pertimbangkan bahwa kita memiliki dua objek yang merupakan representasi fisik dari suatu kelas (tipe data yang ditentukan pengguna) dan kita harus menambahkan dua objek dengan operator biner '+' yang menyebabkan kesalahan, karena kompiler tidak tahu bagaimana cara menambahkan dua objek . Jadi kami mendefinisikan metode untuk operator dan proses itu disebut operator overloading. Kami dapat membebani semua operator yang ada tetapi kami tidak dapat membuat operator baru. Untuk melakukan overloading operator, Python menyediakan beberapa fungsi khusus atau fungsi ajaib yang secara otomatis dipanggil ketika dikaitkan dengan operator tertentu. Sebagai contoh, ketika kita menggunakan + operator, metode ajaib (__add__) secara otomatis dipanggil di mana operasi untuk + operator didefinisikan.
