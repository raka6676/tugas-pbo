__PYGAME__

__A.PENGERTIAN PYGAME__

   adalah modul cross-platform dari Python dirancang untuk membuat game. Modulnya dirancang untuk menjadi sederhana, mudah digunakan, dan menyenangkan, kunci dari idologi Pygame.

__B.CARA INSTALASI PYGAME__

Sebelum menggunakan Pygame ada langkah - langkah yg harus di lakukan agar pygame bisa berjalan, berikut langkah - langkahnya.

1.Buka CMD, kemudian ketikkan “pip”
2.Setelah muncul menu dari pip itu sendiri kemudian ketik “pip install pygame”
3.Kemudian akan otomatis mendownload file instalasi modul pygame. Jika sudah muncul tulisan “Successfully installed pygame-1.9.6” , maka pygame sudah terinstall.
4.Dan Pygame siap di gunakan dengan import pygame
